```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
sber_data = pd.read_csv('data/sber_data.csv')
sber_data.head()
```

```python
display(sber_data.tail())
```

```python
sber_data['sub_area'].nunique()
```

```python
sber_data['price_doc'].max()
```

```python
grouped_data = sber_data.groupby(['ecology'])['price_doc'].last()
grouped_data.plot(
    kind='bar',
    grid=True,
    figsize=(12, 4),
    colormap='plasma'
)
```

```python
fig = plt.figure(figsize=(10, 5))
sns.boxplot(data=sber_data, x='price_doc', y='ecology')
```

```python

fig = plt.figure(figsize=(8, 4))
axes = fig.add_axes([0, 0, 1, 1])
axes.scatter(
    x=sber_data['price_doc'], 
    y=sber_data['kremlin_km'], 
    s=100,
    marker='o',
    c = 'blue'
)
```
