```python
import pandas as pd
melb_data = pd.read_csv('data/melb_data.csv', sep = ',')
display(melb_data.head())
```

```python
melb_data_fe = pd.read_csv('data/melb_data_fe.csv', sep = ',')
display(melb_data_fe.head())
```

```python
display(melb_data_fe.info())
```

```python
melb_data_fe['Date'] = pd.to_datetime(melb_data_fe['Date'])
melb_data_fe['Quarter'] = melb_data_fe['Date'].dt.quarter
display(melb_data_fe.head())
```

```python
display(melb_data_fe['Quarter'].value_counts())
```

```python
display(melb_data_fe.nunique())
```

```python
# 1.2
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_unique_count = 150
for col in melb_data_fe.columns:
    if melb_data_fe[col].nunique() < max_unique_count and col not in cols_to_exclude:
        melb_data_fe[col] = melb_data_fe[col].astype('category')
display(melb_data_fe.info())
```
